```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df.describe()
```

```python
corr = df.corr()
sns.heatmap(
    corr,
    xticklabels=corr.columns,
    yticklabels=corr.columns
)
```

```python
df2 = df[["OverallQual", "GrLivArea", "GarageCars", "GarageArea", "TotalBsmtSF", "SalePrice"]]
df2.head()
```

```python
### Regression LinÃ©aire

y = df.SalePrice
X = df[["OverallQual", "GrLivArea", "GarageCars", "GarageArea", "TotalBsmtSF"]]

xtrain, xtest, ytrain, ytest = train_test_split(X, y)
regr = linear_model.LinearRegression().fit(xtrain, ytrain)

print("Le coefficient Î²1 est Ã©gal Ã  {} tandis que l'intercept Î²0 est Ã©gal Ã  {}"
      .format(round(regr.coef_[0],2),round(regr.intercept_)))

ytrainpred = regr.predict(xtrain)

plt.scatter(xtrain, ytrain, color="red")
plt.plot(xtrain, ytrainpred, color="blue", linewidth=3)
```

```python
### Ridge Regression Linear

regr = linear_model.Ridge(alpha=.5).fit(xtrain, ytrain)

print("Le coefficient Î²1 est Ã©gal Ã  {} tandis que l'intercept Î²0 est Ã©gal Ã  {}"
      .format(round(regr.coef_[0],2),round(regr.intercept_)))

ytrainpred = regr.predict(xtrain)
plt.scatter(xtrain, ytrain, color="red")
plt.plot(xtrain, ytrainpred, color="blue", linewidth=3)
```

```python
### Lasso

regr = linear_model.Lasso(alpha=0.1).fit(xtrain, ytrain)

print("Le coefficient Î²1 est Ã©gal Ã  {} tandis que l'intercept Î²0 est Ã©gal Ã  {}"
      .format(round(regr.coef_[0],2),round(regr.intercept_)))

ytrainpred = regr.predict(xtrain)
plt.scatter(xtrain, ytrain, color="red")
plt.plot(xtrain, ytrainpred, color="blue", linewidth=3)
```
